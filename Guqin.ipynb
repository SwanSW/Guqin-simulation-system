```python
import pygame
import sys
import serial
import numpy as np
import os

# 设置串口参数
port = 'COM7'
baud_rate = 38400

# 初始化串口
s = serial.Serial(port, baud_rate)
# s = "1000,1000,1000,1000,1000,1000,1000,1000"
# 初始化pygame
pygame.init()

# 设置音频通道
pygame.mixer.init()
pygame.mixer.set_num_channels(100)  # 设置音频通道数，例如32个通道

# 设置窗口大小
window_size = (800, 600)
screen = pygame.display.set_mode(window_size)
pygame.display.set_caption('古琴模拟器')

# 定义音频文件目录和文件名前缀
audio_dir = '音频素材'
file_prefixes = [filename[:-4] for filename in os.listdir(audio_dir) if filename.endswith('.wav')]

# 加载音频文件
sounds = {}
for prefix in file_prefixes:
    file_path = os.path.join(audio_dir, f'{prefix}.wav')
    sounds[prefix] = pygame.mixer.Sound(file_path)

# 验证加载的音频文件
#for key, sound in sounds.items():
    #print(f'Loaded sound for key: {key}')

d = {
'010':0,
'020':4.75,
'030':8.55,
'040':14.25,
'050':23.75,
'060':31.35,
'070':42.75,
'080':54.15,
'090':61.75,
'100':71.25,
'110':76.95,
'120':80.75,
'130':85.5,
'000':90
}

def find_closest_key(value, dictionary):
    closest_key = None
    min_diff = float('inf')
    for key, val in dictionary.items():
        diff = abs(value - val)
        if diff < min_diff:
            min_diff = diff
            closest_key = key
    return closest_key

while 1:
    # 读取串口数据
    if s.in_waiting > 0:
        data = s.readline().decode('utf-8').strip()
        new_data = [float(x) for x in data.split(',')]
        if len(new_data) == 3:
            serial_key_right = str(int(new_data[0]))
            serial_value = new_data[1] / 10.0
            serial_key = find_closest_key(serial_value, d)
            print(str(new_data[1])+','+str(new_data[2])+','+str(serial_key))
            if serial_key:
                if new_data[1]==0:
                    data_key = serial_key_right + '0000'
                else:
                    data_key = serial_key_right + str(int(new_data[2])) + serial_key
                print(data_key)
                if data_key in sounds:
                    sounds[data_key].stop()  # 停止当前播放的声音
                sounds[data_key].play()  # 播放新声音 
    # 更新显示
    pygame.display.flip()
    
# 关闭串口和pygame
s.close()
pygame.quit()
sys.exit()

```
